This file compares the results of edge detection using functions coded **from scratch** and **built-in** edge detection functions

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np

from filter.edge_detector_algorithm.canny_edge_detection import canny_edge_detection
from filter.edge_detector_algorithm.marr_hildreth_edge_detection import marr_hildreth_edge_detection
from filter.edge_detector_algorithm.sobel_edge_detection import sobel_edge_detection

sample_image = cv2.imread('samples/selfie.png')
```

**Canny Edge Detection**

```python
canny_scratch = canny_edge_detection(sample_image)

canny_opencv = cv2.Canny(sample_image, 100, 150)

plt.figure(figsize=(20, 30))

plt.subplot(1, 2, 1)
plt.imshow(canny_scratch, cmap='gray')
plt.title('Function coded from scratch', fontsize=30) 
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(canny_opencv, cmap='gray')
plt.title('Built-in function', fontsize=30)
plt.axis('off') 

plt.tight_layout()
plt.show()
```

**Marr Hildreth Edge Detection**

```python
marr_hildreth_scratch = marr_hildreth_edge_detection(sample_image)

gray_image = cv2.cvtColor(sample_image, cv2.COLOR_BGR2GRAY)
laplacian = cv2.Laplacian(gray_image, cv2.CV_64F)
laplacian_abs = np.absolute(laplacian)
marr_hildreth_opencv = cv2.convertScaleAbs(laplacian_abs)

plt.figure(figsize=(20, 30))

plt.subplot(1, 2, 1)
plt.imshow(marr_hildreth_scratch, cmap='gray')
plt.title('Function coded from scratch', fontsize=30) 
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(marr_hildreth_opencv, cmap='gray')
plt.title('Built-in function', fontsize=30)
plt.axis('off')

plt.tight_layout()
plt.show()
```

**Sobel Edge Detection**

```python
sobel_scratch = sobel_edge_detection(sample_image)

gray_image = cv2.cvtColor(sample_image, cv2.COLOR_BGR2GRAY)
grad_x = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0)
grad_y = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1)
sobel_opencv = np.sqrt(grad_x ** 2 + grad_y ** 2)
sobel_opencv = (sobel_opencv * 255 / sobel_opencv.max()).astype(np.uint8)

plt.figure(figsize=(20, 30))

plt.subplot(1, 2, 1)
plt.imshow(sobel_scratch, cmap='gray')
plt.title('Function coded from scratch', fontsize=30) 
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(sobel_opencv, cmap='gray')
plt.title('Built-in function', fontsize=30)
plt.axis('off')

plt.tight_layout()
plt.show()
```
